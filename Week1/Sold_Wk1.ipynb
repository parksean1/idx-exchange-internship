# Sold Week 1

```python
import pandas as pd
import glob

sold_files = sorted(glob.glob('../IDX_Data/CRMLSSold*.csv'))

sold_list = []
sold_rows_before = 0

for file in sold_files:
    df = pd.read_csv(file)
    sold_rows_before += len(df)
    sold_list.append(df)

print("SOLD — Total rows BEFORE concat:", sold_rows_before)
```

```python
sold = pd.concat(sold_list, ignore_index=True)
print("SOLD — Rows AFTER concat:", sold.shape[0])
```

```python
# Check property types
print("SOLD Property Types:", sold['PropertyType'].unique())

# Filter Residential
sold_res = sold[sold['PropertyType'] == 'Residential']
print("SOLD — Rows AFTER Residential filter:", sold_res.shape[0])
```

```python
sold_res.to_csv("sold_combined_residential.csv", index=False)
print("SOLD dataset saved.")
```
