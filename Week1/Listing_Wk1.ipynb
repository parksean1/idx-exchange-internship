# Listing Week 1

```python
import pandas as pd
import glob

listing_files = sorted(glob.glob('../IDX_Data/CRMLSListing*.csv'))

listing_list = []
listing_rows_before = 0

for file in listing_files:
    df = pd.read_csv(file)
    listing_rows_before += len(df)
    listing_list.append(df)

print("LISTING — Total rows BEFORE concat:", listing_rows_before)
```

```python
listing = pd.concat(listing_list, ignore_index=True)
print("LISTING — Rows AFTER concat:", listing.shape[0])
```

```python
# Check property types
print("LISTING Property Types:", listing['PropertyType'].unique())

# Filter Residential
listing_res = listing[listing['PropertyType'] == 'Residential']
print("LISTING — Rows AFTER Residential filter:", listing_res.shape[0])
```

```python
listing_res.to_csv("listing_combined_residential.csv", index=False)
print("LISTING dataset saved.")
```
