# Sold

```python
import pandas as pd
import glob

files = sorted(glob.glob('../IDX Data/CRMLSSold*.csv'))

sold_list = []

for file in files:
    print(f'Loading {file}')
    df = pd.read_csv(file)
    sold_list.append(df)

sold = pd.concat(sold_list, ignore_index=True)

print('Shape:', sold.shape)
```

```python
# Convert to dt objects
date_cols = [
    'CloseDate',
    'ListingContractDate',
    'PurchaseContractDate',
    'ContractStatusChangeDate'
]
for col in date_cols:
    sold[col] = pd.to_datetime(sold[col], errors='coerce')
```
