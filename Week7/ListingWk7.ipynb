# Listing Week 7 

```python
import pandas as pd

listing = pd.read_csv('../../IDX_Data/listing_week6_features.csv')
print('Initial shape:', listing.shape)
```

```python
def get_iqr_bounds(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper
```

```python
# ClosePrice
cp_low, cp_high = get_iqr_bounds(listing['ListPrice'])
listing['listprice_outlier_flag'] = (
    (listing['ListPrice'] < cp_low) | 
    (listing['ListPrice'] > cp_high)
)

# LivingArea
la_low, la_high = get_iqr_bounds(listing['LivingArea'])
listing['livingarea_outlier_flag'] = (
    (listing['LivingArea'] < la_low) | 
    (listing['LivingArea'] > la_high)
)

# DaysOnMarket
dom_low, dom_high = get_iqr_bounds(listing['DaysOnMarket'])
listing['daysonmarket_outlier_flag'] = (
    (listing['DaysOnMarket'] < dom_low) | 
    (listing['DaysOnMarket'] > dom_high)
)
```

```python
print('ListPrice outliers:', listing['listprice_outlier_flag'].sum())
print('LivingArea outliers:', listing['livingarea_outlier_flag'].sum())
print('DaysOnMarket outliers:', listing['daysonmarket_outlier_flag'].sum())
```

```python
listing_filtered = listing[
    (~listing['listprice_outlier_flag']) &
    (~listing['livingarea_outlier_flag']) &
    (~listing['daysonmarket_outlier_flag'])
]

print('Before:', listing.shape[0])
print('After:', listing_filtered.shape[0])
```

```python
print('--- Median Comparison ---')

print('ListPrice BEFORE:', listing['ListPrice'].median())
print('ListPrice AFTER:', listing_filtered['ListPrice'].median())

print('LivingArea BEFORE:', listing['LivingArea'].median())
print('LivingArea AFTER:', listing_filtered['LivingArea'].median())

print('DaysOnMarket BEFORE:', listing['DaysOnMarket'].median())
print('DaysOnMarket AFTER:', listing_filtered['DaysOnMarket'].median())
```

```python
listing.to_csv('../../IDX_Data/listing_week7_flagged.csv', index=False)
listing_filtered.to_csv('../../IDX_Data/listing_week7_filtered.csv', index=False)

print('Week 7 listing datasets saved.')
```
